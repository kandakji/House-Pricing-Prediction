## Task 1 - Data Exploration

### Import needed tools

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import Imputer, LabelEncoder, OneHotEncoder, StandardScaler
````

### Import the dataset

````python
# Base Directory
base = "./challenge_data/"

# Import the Training Data
housePricingDF_train = pd.read_csv(base + 'train.csv',
                                  sep = ',',
                                  header = 'infer',
                                  na_values=['(NA)'])

# Import the Test Data
housePricingDF_test = pd.read_csv(base + 'test.csv',
                                  sep = ',',
                                  header = 'infer',
                                  na_values=['(NA)'])

# Concatenate the Test and Training Data for easier modeling
data = pd.concat([housePricingDF_train, housePricingDF_test], sort = False).drop(columns = ['Id'])

# Store the attributes of the data
features = list(data.columns.values)

# Display the first five entries
data[:5]
````

### Study the data properties and fix some missing values

````python
# Identify the numerical attributes
numeric_features = ['LotFrontage',
                    'LotArea',
                    'YearBuilt',
                    'YearRemodAdd',
                    'MasVnrArea',
                    'BsmtFinSF1',
                    'BsmtFinSF2',
                    'BsmtUnfSF',
                    'TotalBsmtSF',
                    '1stFlrSF',
                    '2ndFlrSF',
                    'LowQualFinSF',
                    'GrLivArea',
                    'BsmtFullBath',
                    'BsmtHalfBath',
                    'FullBath',
                    'HalfBath',
                    'BedroomAbvGr',
                    'KitchenAbvGr',
                    'TotRmsAbvGrd',
                    'Fireplaces',
                    'GarageYrBlt',
                    'GarageCars',
                    'GarageArea',
                    'WoodDeckSF',
                    'OpenPorchSF',
                    'EnclosedPorch',
                    '3SsnPorch',
                    'ScreenPorch',
                    'PoolArea',
                    'MiscVal',
                    'YrSold',
                    'OverallQual',
                    'OverallCond'
                   ]

# Count the number of entries in our data
count = len(data)

# Convert the Sale Prices to numeric
data['SalePrice'] = pd.to_numeric(data['SalePrice'], errors='coerce', downcast='integer')

# Compute the percentage of NA for all numeric attributes
for nf in numeric_features:
    data[nf] = pd.to_numeric(data[nf], errors='coerce', downcast='integer')

    percentage =   (data[nf].isnull().sum(axis = 0))*100/count
    
    if percentage > 0:
        print("Percentage of nulls for",nf,"is", percentage,"%")    
````

````python
fig, ax = plt.subplots()
fig.set_size_inches(12, 6)

# Plot the cumulative distribution
housePricingDF_train["SalePrice"].hist(cumulative=True, density=1, bins='auto',ax=ax)

colors = {0:"red", 1: "orange", 2:"yellow", 3:"darkgreen", 4:"cyan"}

# Calculate and plot the 5 important percentiles
for i, quantile in enumerate([.1, 0.25, .5, .75, .9]):
    ax.axvline(housePricingDF_train["SalePrice"].quantile(quantile), label='%.0fth perc.' % (quantile*100), color = colors[i])
    ax.axhline(quantile, linestyle='-')
    
ax.set_title('ECDF of House Prices')
ax.set_xlabel('House Price')
ax.set_ylabel('ECDF')
ax.legend()
ax.grid(False)
ax.annotate('MEDIAN', (1,0.5), size = 15)
plt.show()
````

````python
# View the relationship between masonry veneer type and masonry veneer area
data[data.MasVnrArea.isnull()][['MasVnrArea', 'MasVnrType']]
````

````python
# Replace the NA's with appropriate values
data.MasVnrArea = data.MasVnrArea.fillna(0)
data.MasVnrType = data.MasVnrType.fillna('None')
````

````python
# Display the relationship between the garage properties to identify how we can fill the missing data
data[data.GarageYrBlt .isnull()][['GarageYrBlt', 'GarageType', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond']]
````

````python
# Replace NA's with the corresponding values
data.GarageYrBlt = data.GarageYrBlt.combine_first(data['YearBuilt'])

# We replaced NA with an 'NA' string to assign it as a category later on
data.GarageType = data.GarageType.fillna('NA')
data.GarageFinish = data.GarageFinish.fillna('NA')
data.GarageQual = data.GarageQual.fillna('NA')
data.GarageCond = data.GarageCond.fillna('NA')
````

````python
# Fill the Lot Frontage missing values with the average
mean = data.LotFrontage.mean()
data.LotFrontage = data.LotFrontage.fillna(mean)
````

````python
non_numeric_features = [f for f in features if f not in numeric_features]
non_numeric_features.remove('SalePrice')

# Compute the percentage of NA for all non-numeric attributes
for nnf in non_numeric_features:
    percentage =   (data[nf].isnull().sum(axis = 0))*100/count
    
    if percentage > 0:
        print("Percentage of nulls for",nnf,"is", percentage,"%")
````

````python
# Drop the columns that have a have proportion of missing values (> 80 %)
data = data.drop(columns = ['Alley', 'PoolQC', 'Fence', 'MiscFeature']) 

non_numeric_features.remove('Alley')
non_numeric_features.remove('PoolQC')
non_numeric_features.remove('Fence')
non_numeric_features.remove('MiscFeature')

features.remove('Alley')
features.remove('PoolQC')
features.remove('Fence')
features.remove('MiscFeature')
````

````python
# We replaced NA with an 'NA' string to assign it as a category later on
data.FireplaceQu = data.FireplaceQu.fillna('NA')
data.BsmtFinType1 = data.BsmtFinType1.fillna('NA')
data.BsmtFinType2 = data.BsmtFinType2.fillna('NA')
data.BsmtExposure = data.BsmtExposure.fillna('NA')
data.BsmtCond = data.BsmtCond.fillna('NA')
data.BsmtQual = data.BsmtQual.fillna('NA')
data.Electrical = data.Electrical.fillna('NA')
````

### Remodeling catagorical data

````python
# Remove the attributes that are categorically comparable (quality features)
non_comparable_catagorical_features = non_numeric_features
non_comparable_catagorical_features.remove('Utilities')
non_comparable_catagorical_features.remove('ExterQual')
non_comparable_catagorical_features.remove('ExterCond')
non_comparable_catagorical_features.remove('BsmtQual')
non_comparable_catagorical_features.remove('BsmtCond')
non_comparable_catagorical_features.remove('BsmtExposure')
non_comparable_catagorical_features.remove('BsmtFinType1')
non_comparable_catagorical_features.remove('BsmtFinType2')
non_comparable_catagorical_features.remove('HeatingQC')
non_comparable_catagorical_features.remove('KitchenQual')
non_comparable_catagorical_features.remove('FireplaceQu')
non_comparable_catagorical_features.remove('GarageFinish')
non_comparable_catagorical_features.remove('GarageQual')
non_comparable_catagorical_features.remove('GarageCond')
````

````python
# Assign labels not non-categorically comparable attributes to assign a number of each catagory
for nnf in non_comparable_catagorical_features:
    labelEnc = LabelEncoder()
    data[nnf] = pd.Series(labelEnc.fit_transform(data[nnf]))
````

````python
# Assign manual labels for attributes whose categories are comparable (higher values => higher quality)

data['Utilities'] = data['Utilities'].replace("AllPub",4)
data['Utilities'] = data['Utilities'].replace("NoSewr",3)
data['Utilities'] = data['Utilities'].replace("NoSeWa",2)
data['Utilities'] = data['Utilities'].replace("ELO",1)

data['BsmtExposure'] = data['BsmtExposure'].replace("Gd",4)
data['BsmtExposure'] = data['BsmtExposure'].replace("Av",3)
data['BsmtExposure'] = data['BsmtExposure'].replace("Mn",2)
data['BsmtExposure'] = data['BsmtExposure'].replace("No",1)
data['BsmtExposure'] = data['BsmtExposure'].replace("NA",0)

data['BsmtFinType1'] = data['BsmtFinType1'].replace("GLQ",6)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("ALQ",5)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("BLQ",4)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("Rec",3)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("LwQ",2)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("Unf",1)
data['BsmtFinType1'] = data['BsmtFinType1'].replace("NA",0)

data['BsmtFinType2'] = data['BsmtFinType2'].replace("GLQ",6)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("ALQ",5)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("BLQ",4)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("Rec",3)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("LwQ",2)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("Unf",1)
data['BsmtFinType2'] = data['BsmtFinType2'].replace("NA",0)

data['GarageFinish'] = data['GarageFinish'].replace("Fin",3)
data['GarageFinish'] = data['GarageFinish'].replace("RFn",2)
data['GarageFinish'] = data['GarageFinish'].replace("Unf",1)
data['GarageFinish'] = data['GarageFinish'].replace("NA",0)

for feat in ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond','HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond']:
    data[feat] = data[feat].replace("Ex",5)
    data[feat] = data[feat].replace("Gd",4)
    data[feat] = data[feat].replace("TA",3)
    data[feat] = data[feat].replace("Fa",2)
    data[feat] = data[feat].replace("Po",1)
    data[feat] = data[feat].replace("NA",0)
````

````python
# Apply dummy encoding to eliminate comparability for the non-comparable catagorical attributes
for nnf in non_comparable_catagorical_features:
    data = pd.concat([data, pd.get_dummies(data[nnf], prefix= nnf)], axis = 1).drop(columns = [nnf])
````

### Scaling numerical features

````python
# Re-assign training and test dataset after modeling
train = data[:len(housePricingDF_train)]
test = data[-len(housePricingDF_test):].drop(columns = ['SalePrice'])
````

````python
# Apply Standardization feature scaling
std_sc = StandardScaler()
train[numeric_features] = pd.DataFrame(std_sc.fit_transform(train[numeric_features]))
test[numeric_features] = pd.DataFrame(std_sc.transform(test[numeric_features]))
````
