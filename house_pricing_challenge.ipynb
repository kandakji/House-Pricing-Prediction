### Import the dataset

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
Now, we import the training and testing data into <b>pandas dataframes</b>. We will concatenate the latter two in order to perform preprocessing on them at the same time.

````python
# Base Directory
base = "./challenge_data/"

# Import the Training Data
housePricingDF_train = pd.read_csv(base + 'train.csv',
                                  sep = ',',
                                  header = 'infer',
                                  na_values=['(NA)'])

# Import the Test Data
housePricingDF_test = pd.read_csv(base + 'test.csv',
                                  sep = ',',
                                  header = 'infer',
                                  na_values=['(NA)'])

# Concatenate the Test and Training Data for easier preprocessing
data = pd.concat([housePricingDF_train, housePricingDF_test], sort = False)

# Store the attributes of the data
features = list(data.columns.values)
features.remove('Id')
# Display the first five entries
data[:5]
````

### Study the data properties and fix the missing values

#### Let's first look at the numerical features

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We will first deal with the numerical featrures and identify the percentage of missing values for these features.
</div>

````python
# Identify the numerical attributes
numeric_features = [f for f in features if np.issubdtype(data[f].dtype, np.number)]
numeric_features.remove('MSSubClass')
numeric_features.remove('MoSold')
numeric_features.remove('SalePrice')


# Count the number of entries in our data
count = len(data)

# Convert the Sale Prices to numeric
data['SalePrice'] = pd.to_numeric(data['SalePrice'], errors='coerce', downcast='integer')

missing_value_percentages = {}

# Compute the percentage of NA for all numeric attributes
for nf in numeric_features:
    data[nf] = pd.to_numeric(data[nf], errors='coerce', downcast='integer')

    percentage =   (data[nf].isnull().sum(axis = 0))*100/count
    
    if percentage > 0:
        print("Percentage of nulls for",nf,"is", percentage,"%")
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We obtain three features with missing values: <b>LotFrontage(17.74%)</b>, <b>GarageYrBlt(5.54%)</b>, <b>MasVnrArea(0.54%)</b>.<br>
Let's first view how the relationship between <b>MasVnrArea</b> and <b>MasVnrType</b> is, to make our decision on how to deal with these features.
</div>

````python
# View the relationship between masonry veneer type and masonry veneer area
data[data.MasVnrArea.isnull()][['MasVnrArea', 'MasVnrType']]
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We can see that there is a direct relationship between <b>MasVnrArea</b> and <b>MasVnrType</b>. Thus, according to the data catalog, we can fill the <b>MasVnrType</b> with <b>'None'</b> and <b>MasVnrArea</b> with <b>0</b>.
</div>

````python
# Replace the NA's with appropriate values
data.MasVnrArea = data.MasVnrArea.fillna(0)
data.MasVnrType = data.MasVnrType.fillna('None')
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
Let's now see the how the different Garage features relate to each other in order to identify how we can deal with the missing numerical values in our data.
</div>

````python
# Display the relationship between the garage properties to identify how we can fill the missing data
data[data.GarageYrBlt .isnull()][['GarageYrBlt', 'GarageType', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond']]
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We can see that the Garage features are related and after looking in the catalog, we can fill the missing values of most with of the features with <b>'NA'</b> and the <b>GarageYrBlt</b> missing values with the value of <b>YearBuild</b>  bacuase it makes most sense than to fill them with 0 or so.
</div>

````python
# Replace NA's with the corresponding values
data.GarageYrBlt = data.GarageYrBlt.combine_first(data['YearBuilt'])

# We replaced NA with an 'NA' string to assign it as a category later on
data.GarageType = data.GarageType.fillna('NA')
data.GarageFinish = data.GarageFinish.fillna('NA')
data.GarageQual = data.GarageQual.fillna('NA')
data.GarageCond = data.GarageCond.fillna('NA')
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
For the <b>LotFrontage</b>, we can fill the missing values with the <b>mean</b> of this features so that these missing values dont add any <b>bias</b> in the data.
</div>

````python
# Fill the Lot Frontage missing values with the average
mean = data.LotFrontage.mean()
data.LotFrontage = data.LotFrontage.fillna(mean)
````

#### Let's now look at the categorical features

````python
non_numeric_features = [f for f in features if f not in numeric_features]
non_numeric_features.remove('SalePrice')

# Compute the percentage of NA for all non-numeric attributes
for nnf in non_numeric_features:
    percentage =   (data[nnf].isnull().sum(axis = 0))*100/count
    
    if percentage > 0:
        print("Percentage of nulls for",nnf,"is", percentage,"%")
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
It is evident that some features are almost empty with a high percentage of missing values like <b>Alley, PoolQC, Fence, MiscFeature </b>. With a percentage <b>above 80%</b>, it is safe to sacrifice those features to reduce the complexity of the model.
</div>

````python
# Drop the columns that have a have proportion of missing values (> 80 %)
data = data.drop(columns = ['Alley', 'PoolQC', 'Fence', 'MiscFeature']) 

# Remove them from the lists of features
non_numeric_features.remove('Alley')
non_numeric_features.remove('PoolQC')
non_numeric_features.remove('Fence')
non_numeric_features.remove('MiscFeature')

features.remove('Alley')
features.remove('PoolQC')
features.remove('Fence')
features.remove('MiscFeature')
````

````python
# We replaced NA with an 'NA' string to assign it as a category later on
data.FireplaceQu = data.FireplaceQu.fillna('NA')
data.BsmtFinType1 = data.BsmtFinType1.fillna('NA')
data.BsmtFinType2 = data.BsmtFinType2.fillna('NA')
data.BsmtExposure = data.BsmtExposure.fillna('NA')
data.BsmtCond = data.BsmtCond.fillna('NA')
data.BsmtQual = data.BsmtQual.fillna('NA')
data.Electrical = data.Electrical.fillna('NA')
````

### Remodeling catagorical data

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We will now study the statistical properties of the catagorical data. We will deal with catagorical data from two perspectives: <b>Comparable</b> and <b>Non-comparable</b><br>
For the <b>comparable</b> features, we will assign an integer for each catagory to keep the <b>preference semantic</b> (for example <b>Good become 4, but Average becomes 3, etc.</b> And thus we will avoid using one-hot encoding to avoid increasing the number of the outcome features due to these catagorical features.
</div>

````python
# Remove the attributes that are categorically comparable (quality features)
non_comparable_catagorical_features = non_numeric_features
non_comparable_catagorical_features.remove('Utilities')
non_comparable_catagorical_features.remove('ExterQual')
non_comparable_catagorical_features.remove('ExterCond')
non_comparable_catagorical_features.remove('BsmtQual')
non_comparable_catagorical_features.remove('BsmtCond')
non_comparable_catagorical_features.remove('BsmtExposure')
non_comparable_catagorical_features.remove('BsmtFinType1')
non_comparable_catagorical_features.remove('BsmtFinType2')
non_comparable_catagorical_features.remove('HeatingQC')
non_comparable_catagorical_features.remove('KitchenQual')
non_comparable_catagorical_features.remove('FireplaceQu')
non_comparable_catagorical_features.remove('GarageFinish')
non_comparable_catagorical_features.remove('GarageQual')
non_comparable_catagorical_features.remove('GarageCond')
````

````python
comparable_catagorical_features = ['Utilities',
'ExterQual',
'ExterCond',
'BsmtQual',
'BsmtCond',
'BsmtExposure',
'BsmtFinType1',
'BsmtFinType2',
'HeatingQC',
'KitchenQual',
'FireplaceQu',
'GarageFinish',
'GarageQual',
'GarageCond']
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
Label encode all the <b>non-comparable catagorical features</b> to assign a number of each catagory.
</div>

````python
# Assign labels not non-categorically comparable attributes to assign a number of each catagory
for nnf in non_comparable_catagorical_features:
    labelEnc = LabelEncoder()
    data[nnf] = pd.Series(labelEnc.fit_transform(data[nnf]))
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue ">

### Linear Regression

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We will start with the <b> linear regression</b> and observe the <b>Mean Square Error</b> along with the <b>R-Squared Matrix</b> that it provides with 10-fold cross validation.
</div>

````python
# Create a Linear Regression Model
linear_reg = LinearRegression()

# Initialize Timer
start = time.time()

# Evaluate the model using 10-fold cross validation
cv_results = cross_validate(estimator = linear_reg,
                            X = train_data,
                            y = sale_prices_sc,
                            scoring = ('r2', 'neg_mean_squared_error'),
                            cv = 10,
                            return_train_score=True)
end = time.time()
dur = end - start

# Store the scores obtained
modelPerformanceDict['ModelName'] = 'LinearRegression'
modelPerformanceDict['TrainMSE'] = -cv_results['train_neg_mean_squared_error'].mean()
modelPerformanceDict['TestMSE'] = -cv_results['test_neg_mean_squared_error'].mean()
modelPerformanceDict['TrainR2'] = cv_results['train_r2'].mean()
modelPerformanceDict['TestR2'] = cv_results['test_r2'].mean()

modelPerformanceDF = modelPerformanceDF.append(modelPerformanceDict, ignore_index=True)

print("Elapsed time = ",dur,"s")
# Print the scores obtained
print("\nAverage Train MSE = ", -cv_results['train_neg_mean_squared_error'].mean())
print("\nStandard Deviation Train MSE = ", -cv_results['train_neg_mean_squared_error'].std())
print("\nAverage Test MSE = ", -cv_results['test_neg_mean_squared_error'].mean())
print("\nStandard Deviation Test MSE = ", -cv_results['test_neg_mean_squared_error'].std())
print("\nAverage Train R-Squared = ", cv_results['train_r2'].mean())
print("\nStandard Deviation Train R-Squared = ", cv_results['train_r2'].std())
print("\nAverage Test R-Squared = ", cv_results['test_r2'].mean())
print("\nStandard Deviation Test R-Squared = ", cv_results['test_r2'].std())
````

## Task 4 & 5 - Parameter Optimization and Model Evaluation

````python
# Create a dataframe to store parameters along with their scores
modelPerformanceDF = pd.DataFrame(columns=['learning_rate', 'n_estimators', 'min_samples_leaf', 'max_depth', 'TrainMSE_Mean', 'TrainMSE_StD', 'TestMSE_Mean', 'TestMSE_StD', 'TrainR2_Mean', 'TrainR2_StD', 'TestR2_Mean', 'TestR2_StD'])
modelPerformanceDict = {}
````

<div class="alert alert-success" style="background-color: AliceBlue; border-color: CornflowerBlue "> 
We will now perform <b>grid search</b> in order to tune some parameters for our selected model which is <b>gradient boosting ensemble regresssion</b>.
</div>

````python
# Create a parameter dictionary that iterates over learning_rate and the number of estimators
param_dict = {'loss' : ['ls'],
              'learning_rate' : [0.1+0.01*i for i in np.arange(5)],
              'n_estimators' : np.arange(150,300,50),
              'subsample': [1.0],
              'criterion' : ['friedman_mse'],
              'min_samples_split' : [2],
              'min_samples_leaf' : [1],
              'min_weight_fraction_leaf' : [0.0],
              'max_depth' : [3],
              'min_impurity_decrease' : [0.0],
              'min_impurity_split' : [None],
              'init' : [None],
              'random_state' : [12],
              'max_features' : [None],
              'alpha' : [0.9],
              'verbose' : [0],
              'max_leaf_nodes' : [None],
              'warm_start' : [False],
              'presort' : ['auto']}

# Identify the scoring metrics
scoring_params = ('r2', 'neg_mean_squared_error')

parameter_grid = list(ParameterGrid(param_dict))

start = time.time()

# Perform GridSearch
for parameter_comb in parameter_grid: 
    
    # Create the Gradient Boosting Ensemble Regressor
    gradBoostReg = GradientBoostingRegressor(**parameter_comb)

    # Evaluate the model using 10-fold cross validation
    cv_results = cross_validate(estimator = gradBoostReg,
                                X = train_data,
                                y = sale_prices_sc,
                                scoring = ('r2', 'neg_mean_squared_error'),
                                cv = 10,
                                return_train_score=True)
    
    # Store the used parameters and the scores obtained
    modelPerformanceDict['learning_rate'] = parameter_comb['learning_rate']
    modelPerformanceDict['n_estimators'] = parameter_comb['n_estimators']
    modelPerformanceDict['min_samples_leaf'] = parameter_comb['min_samples_leaf']
    modelPerformanceDict['max_depth'] = parameter_comb['max_depth']
    modelPerformanceDict['TrainMSE_Mean'] = -cv_results['train_neg_mean_squared_error'].mean()
    modelPerformanceDict['TrainMSE_StD'] = -cv_results['train_neg_mean_squared_error'].std()
    modelPerformanceDict['TestMSE_Mean'] = -cv_results['test_neg_mean_squared_error'].mean()
    modelPerformanceDict['TestMSE_StD'] = -cv_results['test_neg_mean_squared_error'].std()
    modelPerformanceDict['TrainR2_Mean'] = cv_results['train_r2'].mean()
    modelPerformanceDict['TrainR2_StD'] = cv_results['train_r2'].std()
    modelPerformanceDict['TestR2_Mean'] = cv_results['test_r2'].mean()
    modelPerformanceDict['TestR2_StD'] = cv_results['test_r2'].std()

    modelPerformanceDF = modelPerformanceDF.append(modelPerformanceDict, ignore_index=True)

end = time.time()
dur = end - start
print("Grid Search Duration = ",dur)
````
